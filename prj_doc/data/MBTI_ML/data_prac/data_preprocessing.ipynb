```python
import re
import csv

def clean_text(inputString):
  text_rmv = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', ' ', inputString)
  return text_rmv

f = open('train_data_v5_min.csv', 'r', encoding='utf-8-sig')
# wf = open('train_data_v5_clean.csv','a', newline='')
wf = open('train_data_v5_clean.txt','a', newline='')
rdr = csv.reader(f)
for line in rdr:
  for i in line:
    # print(i)
    i_clean = clean_text(i)
    print(i_clean)
    # wr = csv.writer(wf)
    wf.write(i_clean)


wf.close()


f.close()
```
